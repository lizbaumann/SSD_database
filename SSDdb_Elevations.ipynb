### 2. Preprocessing

```python
dfe = dfe1.copy()

dfe.columns = map(str.strip, dfe.columns)
dfe.rename(columns={
    'Transaction Number' : 'TransID',
    'Amount Debit' : 'Debit',
    'Amount Credit' : 'Credit',
    'Description' : 'El_Description',
    'Memo' : 'El_Memo',
    'Check Number' : 'El_CheckNbr'
    }, inplace=True)
cols = ['Date', 'SourceFile', 'Account', 'TransID',
    'Debit', 'Credit', 'Balance', 
    'El_Description', 'El_Memo', 'El_CheckNbr']
dfe = dfe[cols]
dfe[['El_Description']] = dfe[['El_Description']].astype(str)
dfe[['El_Memo']] = dfe[['El_Memo']].astype(str)
dfe['El_CheckNbr'] = dfe['El_CheckNbr'].fillna(0.0).astype(int).astype(str)
dfe[['Debit']] = dfe[['Debit']].astype(float)
dfe['Debit'].fillna(0, inplace=True)
dfe[['Credit']] = dfe[['Credit']].astype(float)
dfe['Credit'].fillna(0, inplace=True)
dfe['Amount'] = dfe['Debit'] + dfe['Credit']
dfe['Date'] = pd.to_datetime(dfe['Date'], format='%m/%d/%Y')
dfe['For Date'] = dfe['Date']
dfe['For Month'] = dfe['For Date'].apply(lambda dt: int(dt.strftime('%Y%m')))
dfe['Entries'] = 1
#dfe_pre2013 = dfe[dfe['Date'] < '01-01-2013']
#dfe = dfe[dfe['Date'] >= '01-01-2013']

dfe2 = dfe.copy()
dfe.shape # (482, 14)
```

### 3. Assign who, what3 categories

### 'how' in the expenses detail import:
- cash: only item is 16.25 reimburse exp... see how/where this shows up in paypal and elevations
- check and debit card: should be able to merge this to get categories

```python
dfe = dfe2.copy()

def assign_cats_el(s):
    '''Assign descriptors for every transaction:
    who (Paypal, Square, member, vendor, etc.)
    what1: what2: what3: . . . '''
    
    who = 'UNKNOWN'
    what3 = 'UNKNOWN'
    
    ########## Dues, Donations, Dividends, some Fees ##########
    if ('SQUARE' in s['El_Memo'].upper()) | \
        ('SQC' in s['El_Memo'].upper()) | \
        ('SQUARE' in s['El_Description'].upper()):
        who = 'Square'
        what3 = 'Dues and Donations'
    elif ('PAYPAL' in s['El_Memo'].upper()) | \
        ('PAYPAL' in s['El_Description'].upper()):
        who = 'Paypal'
        if s['Amount'] > 0:
            what3 = 'Transfers'
        else:
            what3 = 'Fees Paypal Monthly'
    elif ('PP' in s['El_Memo'].upper()) & \
        (s['El_Description'] == 'Withdrawal'):
        who = 'Paypal'
        what3 = 'Fees Paypal Other'
    elif 'DEPOSIT ADJUSTMENT' in s['El_Description'].upper():
        who = 'Geoffrey Terrell'
        what3 = 'Dues Monthly'
    elif 'HOME BANKING' in s['El_Description'].upper():
        who = 'Self'
        what3 = 'Transfers'
    elif 'DIVIDEND' in s['El_Description'].upper():
        who = 'Elevations'
        what3 = 'Dividends'
    elif 'DEPOSIT' in s['El_Description'].upper():
        what3 = 'Dues and Donations'
    
    ########## Fees ##########
    elif s['El_Description'] == 'Business Fee':
        who = 'Elevations'
        what3 = 'Fees Bank Monthly'
    elif (s['El_Description'] == 'Courtesy Pay Fee') | \
        ('ITEM FEE STALE DATE' in s['El_Memo'].upper()):
        # Courtesy pay fee for? not sure why
        who = 'Elevations'
        what3 = 'Fees Bank Other'
    
    ########## Rent and Utilities ##########
    elif (s['El_Description'] == 'Withdrawal by Check') & \
        (s['For Month'] >= 201202) & (s['For Month'] <= 201204):
        who = 'Westland'
        what3 = 'Rent'
    elif ('Draft' in s['El_Description']) & (s['For Month'] <= 201303) & \
        (s['Amount'] == -1250):
        who = 'Westland'
        what3 = 'Rent'
    elif s['El_CheckNbr'] in ['127', '157', '158', '159', '177', '179']:
        who = 'Westland'
        what3 = 'Utilities'
    elif s['El_CheckNbr'] in ['151', '180', '181', '182', '183']:
        who = 'Westland'
        what3 = 'Rent and Utilities'
    
    elif (s['El_CheckNbr'] == '184') | \
        ((s['Date'] == '04/03/2014') & (s['Amount'] == -61.46)):
        who = 'Kreizel'
        what3 = 'Rent'
    elif (s['El_CheckNbr'] == '227') | \
        (('KREIZEL' in s['El_Description'].upper()) & (s['For Month'] == 201406)):
        who = 'Kreizel'
        what3 = 'Utilities'
    elif (s['El_Description'] in ['226', '228', '229', '231']) | \
        ('KREIZEL' in s['El_Description'].upper()) | \
        ('KREIZEL' in s['El_Memo'].upper()):
        who = 'Kreizel'
        what3 = 'Rent and Utilities'
    
    ########## Internet, Trash ##########
    elif ('LIVE WIRE' in s['El_Memo'].upper()) | \
        ('LIVE WIRE' in s['El_Description'].upper()) | \
        (s['El_CheckNbr'] == '101'):
        who = 'Live Wire'
        what3 = 'Internet'
    elif 'WESTERN DISPOSAL' in s['El_Description'].upper():
        who = 'Western Disposal'
        what3 = 'Trash'
    
    ########## Special Items ##########
    elif s['El_CheckNbr'] == '152':
        who = 'Mill'
        what3 = 'Mill'
    elif s['El_CheckNbr'] == '202':
        who = 'John English'
        what3 = 'Security Deposit Expense'    
    
    ########## Misc Expenses ##########
    elif 'Withdrawal' in s['El_Description']:
        if 'ITEM STALE DATE' in s['El_Memo'].upper():
            who = 'Zooko'
            what3 = 'Dues Monthly'
        elif 'SOS REGISTRATION' in s['El_Memo'].upper():
            who = 'CO Sec of State'
            what3 = 'SOS Registration'
                
    return pd.Series({
        'who' : who,
        'what3' : what3})


# dfe = dfe2.copy()
dfe_cats = dfe.apply(assign_cats_el, axis=1)
dfe = dfe.join(dfe_cats)

dfe3 = dfe.copy()
dfe.shape # (482, 16)
```

### 4. Split what3 = 'Rent and Utilities' 

```python
dfe = dfe3.copy()

# ? try? df.append(s, ignore_index=True)

#dfe2[dfe2['what3'] == 'Rent and Utilities'][['Amount','Entries']].groupby(dfe2['El_Description']).sum()
#dfe2[dfe2['what3'] == 'Rent and Utilities'][['Amount','Entries']].groupby(dfe2['Date']).sum()

def split_rentutil(s):
    if s['For Month'] < 201304:
        rent = -1250
        utilities = s['Amount'] - rent
    elif s['For Month'] < 201308:
        utilities = -150
        rent = s['Amount'] - utilities
    elif s['Amount'] > -200:
        utilities = 0
        rent = s['Amount']
    else:
        utilities = -200
        rent = s['Amount'] - utilities
    return pd.Series({'Rent': rent, 'Utilities': utilities})

dfe_not_ru = dfe[dfe['what3'] != 'Rent and Utilities']

dfe_rent = dfe.copy()
dfe_rent = dfe_rent[dfe_rent['what3'] == 'Rent and Utilities']
dfe_rent['what3'] = 'Rent'
dfe_rent['Amount'] = dfe_rent.apply(split_rentutil, axis=1)['Rent']

dfe_util = dfe.copy()
dfe_util = dfe_util[dfe_util['what3'] == 'Rent and Utilities']
dfe_util['what3'] = 'Utilities'
dfe_util['Amount'] = dfe_util.apply(split_rentutil, axis=1)['Utilities']

dfe = pd.concat([dfe_not_ru, dfe_rent, dfe_util])

#dfe_rent[['Date','what3','who','Amount']].sort('Date')
#dfe_util[['Date','what3','who','Amount']].sort('Date')

dfe4 = dfe.copy()
dfe.shape # (504, 16)
```

```python
# assign less straightforward or derived columns
def assign_dddtl(s):
	''' Assign dues / donations detail fields. Note set it up
	so that payments for multiple months are in separate rows.'''
	Attendees = 0
	Dues_Rate = 0
	Mbrs = 0
	Mbrs_Reg = 0
	Mbrs_SS = 0
	Mbrs_Fam = 0
	Mbrs_UNK = 0
	Dues_Disc = 0
	what3 = s['Category']
	
	if s['For Month'] == 0: 
		For_Month = int(s['Date'].strftime('%Y%m'))
	else: 
		For_Month = s['For Month']
	
	if s['Category'] == 'Workshops':
		Attendees = max(1,s['Qty'])
	elif s['Category'] == 'Dues Monthly':
		Mbrs = 1
		if (s['For Month temp'] == 201412) & (s['who'] == 'John West'):
			Mbrs_SS = 0.5 # 12/1/14 dues split over 2 months
			Mbrs = 0.5
			Dues_Rate = 25.0
		elif ((s['Amount'] >= 11) & (s['Amount'] <= 13)) | \
			((s['Amount'] >= 23) & (s['Amount'] <= 25)) | \
			(s['Amount'] == 30) | (s['Amount'] == 40):
			Mbrs_SS = 1
			Dues_Rate = 25.0
		elif ((s['Amount'] >= 35) & (s['Amount'] <= 38)) | \
			((s['Amount'] >= 62) & (s['Amount'] <= 65)) | \
			((s['Amount'] >= 71) & (s['Amount'] <= 75)):
			Mbrs_Reg = 1
			if (s['Amount'] >= 62) & (s['Amount'] <= 65):
				Dues_Rate = 65.0
			else:
				Dues_Rate = 75.0
		elif ((s['Amount'] >= 48) & (s['Amount'] <= 50)) | \
			((s['Amount'] >= 95) & (s['Amount'] <= 100)):
			Mbrs_Fam = 1
			Dues_Rate = 100.0
		else:
			Mbrs_UNK = 1
		if (Mbrs == 1) & \
			(((s['Amount'] >= 11) & (s['Amount'] <= 13)) | \
			((s['Amount'] >= 31) & (s['Amount'] <= 33)) | \
			((s['Amount'] >= 35) & (s['Amount'] <= 38)) | \
			((s['Amount'] >= 46) & (s['Amount'] <= 50))):
			Dues_Disc = 1
	
	return pd.Series({
		'what3' : what3,
		'For Month' : For_Month,
		'Attendees' : Attendees, 
		'Dues_Rate': Dues_Rate,
		'Mbrs' : Mbrs,
		'Mbrs_Reg' : Mbrs_Reg,
		'Mbrs_SS' : Mbrs_SS,
		'Mbrs_Fam' : Mbrs_Fam,
		'Mbrs_UNK' : Mbrs_UNK,
		'Dues_Disc' : Dues_Disc})


dfe_subs3b = dfe_subs3.apply(assign_dddtl, axis=1)
dfe_subs4 = dfe_subs3.join(dfe_subs3b)

dfe_subs4['Amount'].sum() # 7907.35
```
