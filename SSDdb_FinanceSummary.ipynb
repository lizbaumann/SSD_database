## 2. Get what1, what2 higher level categories

```python
dfall = dfall1.copy()
dfall = dfall.reset_index()
dfall.drop('index', axis=1, inplace=True)

def assign_what12(s):
    '''Assign what1, what2 descriptors for every transaction'''
    
    what1 = 'UNKNOWN'
    what2 = 'UNKNOWN'
    
    if 'Dues' in s['what3']:
        what2 = 'Dues'
        what1 = 'Revenue'
    elif s['what3'] in ['Workshops', 'Donations', 'Dividends']:
        what2 = 'Other Revenue'
        what1 = 'Revenue'
    elif s['what3'] in ['Security Deposit Revenue', '501c3 Fund']:
        what2 = 'Special Revenue'
        what1 = 'Revenue'
    elif s['what3'] == 'Transfers':
        what2 = 'Tranfers'
        what1 = 'Revenue'
    
    elif s['what3'] in ['Rent and Utilities', \
        'Rent', 'Utilities', 'Internet', 'Trash']:
        what2 = 'Rent and Utilities'
        what1 = 'Expenses'
    elif s['what3'] in ['Insurance', 'Taxes', 'SOS Registration', \
        'Fees Bank Monthly', 'Fees Bank Other', \
        'Fees Paypal Monthly', 'Fees Paypal Transactions', \
        'Fees Square']:
        what2 = 'Insurance, Taxes and Fees'
        what1 = 'Expenses'
    elif s['what3'] in ['Consumables', 'Equipment', 'Promotional', \
        'Other Expense']:
        what2 = 'Misc Expenses'
        what1 = 'Expenses'
    elif s['what3'] in ['Security Deposit Expense', 'Mill']:
        what2 = 'Special Expenses'
        what1 = 'Expenses'
    
    return pd.Series({
        'what1' : what1, 
        'what2' : what2})

dfall_what12 = dfall.apply(assign_what12, axis=1)
dfall = dfall.join(dfall_what12)
dfall.shape
```
