## 3. Assign what3 categories
+ 	what1: what2: what3: 
+ 	Revenue: Dues: Dues Monthly, Dues Recurring
+ 	Revenue: Other Revenue: Workshops, Donations, Dividends
+ 	Revenue: Special Revenue: Security Deposit Revenue, 501c3 Fund
+ 	Revenue: Transfers: Transfers
+ 	Expenses: Rent and Utilities: Rent, Utilities, Internet, Trash
+ 	Expenses: Insurance, Taxes and Fees: Insurance, Taxes, SOS Registration, Fees Bank Monthly, Fees Bank Other, Fees Paypal Monthly, Fees Paypal Transactions
+ 	Expenses: Misc Expenses: Consumables, Equipment, Promotional, Other Expense
+ 	Expenses: Special Expenses: Mill, Security Deposit Expense
+ 	TEMPORARY: Dues and Donations, Rent and Utilities


```python
dfp = dfp2.copy()

def assign_what3_pp(s):
    what3 = 'UNKNOWN'
    mult = 1.0
    
    if s['who'] == 'Paypal':
        what3 = 'Fees Paypal Transactions'
    elif (s['who'] == 'Dba Scott K. Moore') & (int(s['Date'].strftime('%Y%m%d')) == 20130622):
        what3 = 'Workshops'
    elif (s['who'] == 'Ryan Bennett') & (int(s['Date'].strftime('%Y%m%d')) == 20150127):
        what3 = 'Dues Monthly'
    elif (s['who'] == 'John Maushammer') & (int(s['Date'].strftime('%Y%m%d')) == 20120808):
        what3 = 'Dues Monthly'
    elif s['PP_Type'] == 'Web Accept Payment Received':
        if 'per person' in s['PP_Item Title']:
            what3 = 'Workshops'
        elif (s['PP_Option 1'] == 'Student/Strapped') | \
            (s['PP_Option 1'] == 'Feb Prorated') | \
            ('Month' in s['PP_Option 1']) | \
            ('Monthly Dues' in s['PP_Item Title']):
            what3 = 'Dues Monthly'
    elif (s['PP_Type'] == 'Recurring Payment Received') | \
        (s['PP_Type'] == 'Subscription Payment Received') | \
        (s['PP_Type'] == 'Payment Received'):
        what3 = 'Dues Recurring'
    elif (s['PP_Type'] == 'Donation Received') | \
        (s['PP_Type'] == 'Mobile Payment Received'):
        what3 = 'Donations'
    elif s['PP_Type'] == 'Refund':
        # usually for workshop discount but just a couple exceptions
        what3 = 'Dues Recurring' # should offset dues payment
    elif s['who'] == 'Bank Account':
        what3 = 'Transfers'
    elif 'Monthly Billing' in s['who']:
        what3 = 'Fees Paypal Monthly'
    
    # multiplier, used for member dues and workshop attendees
    if ('$30.00 per person' in s['PP_Item Title']) & \
        ('3D' in s['PP_Item Title']):
        mult = 2.0
    elif '3 Months' in s['PP_Option 1']:
        mult = 3.0
    elif ('2 Months' in s['PP_Option 1']) & (s['who'] != 'Ryan Bennett'):
        mult = 2.0
    elif ('1 Half-Month' in s['PP_Option 1']) | \
        ('after the 15th' in s['PP_Option 1']) | \
        (s['PP_Option 1'] == 'Feb Prorated'):
        mult = 0.5
    
    return pd.Series({'what3' : what3, 'mult' : mult})


# dfp = dfp1.copy()
dfp_what3 = dfp.apply(assign_what3_pp, axis=1)
dfp = dfp.join(dfp_what3)

dfp3 = dfp.copy()
dfp.shape # (3112, 18)
```

## 8. Derive members, dues rate, dues discount indicator

```python
dfp = dfp7.copy()

def assign_members_pp(s):
    ''' Derive members, dues rate, workshop discount,
    workshop attendees. Prerequisite: what3 is assigned '''
    attendees = 0
    mbrs = 0.0
    duesrate = float(0)
    mbrs_reg = 0.0
    mbrs_ss = 0.0
    mbrs_fam = 0.0
    mbrs_unk = 0.0
    duesdisc = 0
    
    if s['what3'] == 'Workshops':
        if (s['who'] == 'Dba Scott K. Moore') & (s['For Month'] == 201306):
            attendees = 14  # for3d printing
        else:
            attendees = s['mult']
    
    elif 'Dues' in s['what3']:
        if s['mult'] > 1:
            mbrs = 1
        else: 
            mbrs = s['mult']
    
        # dues discount for workshop: either charged half or refunded
        if s['fam_ind'] == 'Y': 
            mbrs_fam = mbrs
            # special family rates, Seb and Kerry at 40/25, Rich and Jordan at 25ea; no discounts
            if s['who'] == 'Sebastian Kuzminsky':
                duesrate = 40
            elif s['who'] == 'Kerry Kruempelstaedter':
                duesrate = 25
            elif s['who'] in ('Richard Buchman', 'Jordan Buchman'):
                duesrate = 25
            else: 
                duesrate = 50
                if s['Amount'] == 25.0:
                    duesdisc = 1
        else:
            if (s['Amount'] == 12.5) | (s['Amount'] == 32.5):
                duesdisc = 1
                duesrate = s['Amount'] * 2.0 / mbrs
            elif (s['Amount'] == 55) & (s['For Month'] < 201301):
                # used to give $10 off, in 2012
                duesdisc = 1
                duesrate = 65.0
            else:
                duesrate = s['Amount'] * 1.0 / mbrs
            
            # EXCEPTIONS: Adjust duesrate for various reimbursements
            if (duesrate == 40.0):
                # a few people have paid 40, count as Student Senior
                duesrate = 25.0 
            elif (duesrate == 50.0):
                # one+ joined end of month and I allocated 25/50, count 50 as Regular
                duesrate = 65.0 
            elif (duesrate >= 75.0) & (duesrate < 77):
                duesrate = 75.0
            elif (duesrate >= 60.0) and (duesrate <= 65.0):
                # Feb prorated in 201202 becomes 62.08
                duesrate = 65.0
            
            if s['Amount'] == 0:
                mbrs = 0
            
            if (duesrate == 65.0) | (duesrate == 75.0):
                mbrs_reg = mbrs
            elif (duesrate == 25.0):
                mbrs_ss = mbrs
            else: 
                mbrs_unk = mbrs
    
    return pd.Series({'Mbrs' : mbrs, 
        'Mbrs_Reg' : mbrs_reg,
        'Mbrs_SS' : mbrs_ss,
        'Mbrs_Fam' : mbrs_fam,
        'Mbrs_UNK' : mbrs_unk,
        'Dues_Rate' : duesrate,
        'Attendees' : attendees,
        'Dues_Disc' : duesdisc})


dfp_mbrs = dfp.apply(assign_members_pp, axis=1)
dfp = dfp.join(dfp_mbrs)

# repopulate columns
dfp['Account'] = 'Paypal'
dfp['For Date'] = dfp['For Month'].apply(lambda x: datetime.date(int(x/100), x - int(x/100)*100, 1))
#dfp['Date'] = dfp['For Date']

# cleanup: fill in nas, keep only necessary well-populated columns
keepcols = ['For Month', 'For Date', 'who', 'what3', 'Account', 
            'Debit', 'Credit', 'Amount', 'Entries', 'Attendees', 
            'Dues_Disc', 'Dues_Rate', 'Mbrs', 'Mbrs_Reg', 'Mbrs_SS', 'Mbrs_Fam', 'Mbrs_UNK']
dfp = dfp[keepcols]


dfp8 = dfp.copy()
dfp.shape # (1373, 17)
```
